```python
def scrap_stock_info(request_dict):
    response_dict = {}
    for name, data in request_dict.items():
        url = data['url']
        params = data['params']
        response = requests.get(url=url,params=params)
        json_data = json.loads(response.text)
        response_dict[name]=json_data

    return response_dict
```

```python
response_dict = scrap_stock_info(request_dict)
```

```python
parse_dict = {
    'binance':{'bids':'bids', 'asks':'asks'},
    'okex':{'bids':'bids', 'asks':'asks'},
    'bitfinex':'Special Case',
    'huobi':{'bids':['tick','bids'], 'asks':['tick','asks']},
    'bithumb':{'bids':['data','b'], 'asks':['data','s']},
    'hitbtc':{'bids':'bid', 'asks':'ask'},
    'kraken':{'bids':['result','XBTUSDT','bids'], 'asks':['result','XBTUSDT','asks']},
    'coinbasepro':{'bids':'bids', 'asks':'asks'},
    'bitstamp':{'bids':'bids', 'asks':'asks'},
    'gemini':{'bids':'bids', 'asks':'asks'}
}
```

```python
def parse_bidask(response_dict, parse_dict):
    parsed_bidask = {}
    for name, labels in parse_dict.items():
        if labels == 'Special Case':
            bids = response_dict.get(name)[:25]
            asks = response_dict.get(name)[25:]
            parsed_bidask[name] = {'bids': bids,'asks':asks}
            continue
            
        data = response_dict.get(name)
        bids_label = labels['bids']
        asks_label = labels['asks']
        
        if type(bids_label) is list:
            bids = data
            for level in bids_label:
                bids = bids.get(level)
        else:
            bids = data.get(bids_label)
        
        if type(asks_label) is list:
            asks = data
            for level in asks_label:
                asks = asks.get(level)
        else:
            asks = data.get(asks_label)
            
        parsed_bidask[name] = {'bids': bids,'asks':asks}

    return parsed_bidask
```

```python
parsed_bidask = parse_bidask(response_dict, parse_dict)
```

```python
clean_dict = {
    'binance':['price', 'amount'],
    'okex':['price', 'amount', 'num_orders'],
    'bitfinex':['price', 'num_orders', 'amount'],
    'huobi':['price', 'amount'],
    'bithumb':['price', 'amount'],
    'hitbtc':{'price':'price', 'amount':'size'},
    'kraken':['price', 'amount','timestamp'],
    'coinbasepro':['price', 'amount', 'num_orders'],
    'bitstamp':['price', 'amount'],
    'gemini':{'price':'price', 'amount':'amount','timestamp':'timestamp'}
}
```

```python
def clean_bidask(parsed_bidask, clean_dict, needed_labels):
    cleared_bidask = {}
    for name, labels in clean_dict.items():
        cleared_entry = {'bids':[],'asks':[]}
        
        for market_marker, data in parsed_bidask[name].items():
        
            for row in data:
                entry = {}
                for label in needed_labels:
                    entry[label] = 0.0
                if type(data[0]) is list:
                    for counter, label in enumerate(labels):
                        if label in needed_labels:
                            entry[label] = row[counter]

                elif type(data[0]) is dict:
                    for mname, subname in labels.items():
                        if mname in needed_labels:
                            entry[mname] = row[subname]
                entry['source']=name
                cleared_entry[market_marker].append(entry)               
                
        cleared_bidask[name] = cleared_entry

    return cleared_bidask
```

```python
def get_bidasks(cleared_bidask):
    bidask = {'bids':[], 'asks':[]}
    for name, data in cleared_bidask.items():
        bids, asks = data['bids'], data['asks']
        bidask['bids'] += bids 
        bidask['asks'] += asks
    bids, asks = [], []
    for entry in bidask['bids']:
        price, amount,name = entry['price'], entry['amount'], entry['source']
        bids.append((float(price), float(amount), name))
    for entry in bidask['asks']:
        price, amount, name = entry['price'], entry['amount'], entry['source']
        asks.append((float(price), float(amount), name))  
        
    return bids, asks
```

```python
def calculate_vwap(bands, bids, asks):
    
    def calculate(bands, smth):
    
        vwap_by_band = {}
        total = 0
        amount = 0
        bids = 0
        band_pointer = 0
        band = bands[band_pointer]
        counter = 0
        max_band_ptr = len(bands) - 1

        for entry in smth:
            price, amount, _ = entry
            total += price * amount
            amount += amount
            counter += 1
            while amount > band and band_pointer <= max_band_ptr:
                vwap = total / counter
                vwap_by_band[band] = vwap
                if band_pointer == max_band_ptr:
                    break
                if band_pointer < max_band_ptr:
                    band_pointer += 1
                    band = bands[band_pointer]
        return vwap_by_band
    
    bids = sorted(bids, key=lambda x: x[0], reverse=True)
    asks = sorted(asks, key=lambda x: x[0], reverse=False)
    
    bids_vwap = calculate(bands, bids)
    asks_vwap = calculate(bands, asks)
    
    return bids_vwap, asks_vwap
        
```
